```python
# import libraries
from scapy.all import *
from collections import Counter
from prettytable import PrettyTable
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from datetime import datetime
import pandas as pd
from pcap_processor import processor as proc

init_notebook_mode(connected=True)
```

```python
# Read packet capture into a variable
packets = rdpcap('sample.pcap')
```

```python
# Initialize a source IP list and read through packets storing source IP headers
srcIP=[]
pktBytes=[]
pktTimes=[]
for pkt in packets:
    if IP in pkt:
        # try and except block is used to iterate over possible malformed information
        try:
            srcIP.append(pkt[IP].src)
            pktBytes.append(pkt[IP].len)
            pktTime=datetime.fromtimestamp(pkt.time)
            pktTimes.append(pktTime.strftime("%Y-%m-%d %H:%M:%S.%f"))
        except:
            pass
```

```python
# Count each occurence of source IP address
cnt=Counter()

for ip in srcIP:
    cnt[ip] += 1
```

```python
# Create table and populate data to display
table= PrettyTable(["IP", "Count"])

for ip, count in cnt.most_common():
   table.add_row([ip, count])
```

```python
print(table)
```

```python
# Prepare data for charting
xData=[]
yData=[] 

for ip, count in cnt.most_common():
    xData.append(ip)
    yData.append(count)
    
plotly.offline.iplot({"data":[  plotly.graph_objs.Bar( x=xData, y=yData) ] })
```

```python
#This converts list to series

bytes = pd.Series(pktBytes).astype(int)

#Convert the timestamp list to a pd date_time
times = pd.to_datetime(pd.Series(pktTimes).astype(str),  errors="coerce")

#Create the dataframe
df  = pd.DataFrame({"Bytes": bytes, "Times":times})

#set the date from a range to an timestamp
df = df.set_index("Times")

df2=df.resample("2S").sum()
print(df2)
```

```python
#Create the graph
plotly.offline.iplot({"data":[plotly.graph_objs.Scatter(x=df2.index, y=df2["Bytes"])],
"layout":plotly.graph_objs.Layout(title="Bytes over Time ",
xaxis=dict(title="Time"),
yaxis=dict(title="Bytes"))})
```
